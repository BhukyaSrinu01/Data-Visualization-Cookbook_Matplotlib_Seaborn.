```python
#importing basic libraries essential for our code...

import matplotlib.pyplot as plt  #importing matplot library using ".pyplot" used for making graphs and plots

import numpy as np  # numpy is used for arrays and operations to be performed on them .

```

```python
# Randomly sampled data
x = np.random.normal(0, 1, 1000)# The code is used to generate a random number between 0 and 1,000 for x
y = np.random.normal(0, 1, 1000)# The code is used to generate a random number between 0 and 1,000 for y.
z = np.random.normal(0, 1, 1000)#  The code is used to generate a random number between 0 and 1,000 for z.

labels = ['A', 'B', 'C', 'D']# A list having a set of strings.
values = [25, 35, 28, 9, 3]# A list having a set of Integers.
```

# 1.Histogram :
A histogram is a graphical representation of data points organized into user-specified ranges. Similar in appearance to a bar graph, the histogram condenses a data series into an easily interpreted visual by taking many data points and grouping them into logical ranges or bins . A histogram displays numerical data by grouping data into "bins" of equal width. Each bin is plotted as a bar whose height corresponds to how many data points are in that bin.Bins are also sometimes called "intervals", "classes", or "buckets".

```python
# creating a Histogram

# plt.hist() is a function of pyplot used to create a Histogram 
plt.hist(x, bins=30)

#setting the x label as "value"
plt.xlabel('Value')

#setting the y label to be "Frequency".
plt.ylabel('Frequency')

#setting the title of the graph as "Histogram"
plt.title('Histogram')

#plt.show is used to show the plotted functions above..
plt.show()
```

# 2 - Barchart :
A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value. Some bar graphs present bars clustered in groups of more than one, showing the values of more than one measured variable

```python
# import matplotlib.pyplot as plt

# Sample data
labels = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Check if the length of labels and values is the same
if len(labels) == len(values):
    # Creating a Barchart
    plt.bar(labels, values)

    # Labelling the title of x-axis as "Category".
    plt.xlabel('Category')
    plt.show()
else:
    print("Error: The length of labels and values should be the same.")
```

# Barchart vs Histogram.

Both histograms and bar charts provide a visual display using columns, and people often use the terms interchangeably. Technically, however, a histogram represents the frequency distribution of variables in a data set. A bar graph typically represents a graphical comparison of discrete or categorical variables.

# 3 - PieChart :

A pie chart, sometimes called a circle chart, is a way of summarizing a set of nominal data or displaying the different values of a given variable (e.g. percentage distribution). This type of chart is a circle divided into a series of segments. Each segment represents a particular category

```python
# A pie chart is created using plt.pie() function in pyplot.
# In the below statetment we declared values from the list of "values" and choses the list of "labels" to label them.

plt.pie(values, labels=labels)

## We use plt.title to set the label of any particular graph..
plt.title('A Pie or a Circular Chart')

## plt.show is used for displaying the graph.
plt.show()
```

# 4 - Box-Plot :
A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
# plt.boxplot() is a function used to initialize a boxplot with three variables as x and y
plt.boxplot([x,y])

## Labelling X-axis to be "Variable".
plt.xlabel('Varible')

## Labelling the Y-axis as "Value"
plt.ylabel('Value')

##Label the title of graph as "Box Plot".
plt.title('Box Plot')

##Last but not Least the plt.show() to display your graph
plt.show()
```

# 5 - Scatterplot:
Scatter plots are the graphs that present the relationship between two variables in a data-set. It represents data points on a two-dimensional plane or on a Cartesian system. The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis. These plots are often called scatter graphs or scatter diagrams.

```python
# use plt.scatter( ) to display a scatterplot.
plt.scatter(x, y)
# Plot the label of X-axis as "X".
plt.xlabel('X')
# Plot the label of Y-axis as "Y".
plt.ylabel('Y')
#Label the Title of graph as "Scatter Plot".
plt.title('Scatter Plot')
# Using plt.show() to display the graph AGAIN...
plt.show()
```

# 9 - WordCloud:
A word cloud is a visual representation of text data where the size of each word corresponds to its frequency or importance within the text. It is a popular data visualization technique used to quickly grasp the key themes, keywords, or trends present in a body of text. Word clouds provide an intuitive and visually appealing way to highlight the most prominent words in a text corpus.

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#Sample data(replace with your own data)
text = 'data analytics visualization machine learning python kaggle data science'

#Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

#create and configure the plot
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

#add title
plt.title('Word Cloud')

#Display the plot
plt.show()
```

# 11 - 3D_Scatterplot:
A 3D scatter plot is a type of visualization that represents data points in a three-dimensional space. It is used to explore relationships between three numerical variables. In a 3D scatter plot, each data point is plotted using three coordinates (x, y, z) on a three-dimensional Cartesian plane.

```python
# Generate random data
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y , z)

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot')

# Display the plot
plt.show()
```

# 12 - Stacked Area Plot:
A stacked area plot is a type of visualization that displays the cumulative contribution of multiple variables over a continuous axis, typically represented as a line or an area chart. In a stacked area plot, each variable's value is stacked on top of the previous variable, creating a cumulative representation of their combined contribution.

```python
#data
years = [2010, 2011, 2012, 2013, 2014, 2015]
income = [25000, 35000, 45000, 55000, 50000, 60000]
expenses = [22000, 28000, 35000, 40000, 42000, 50000]

#create a stacked area plot
plt.stackplot(years, income, expenses, labels=['Income', 'Expenses'])

#add labels and title
plt.xlabel('years')
plt.ylabel('Amount')
plt.title('Stacked Area Plot')

#Add lengend
plt.legend()

#Display the plot
plt.show()
```

# 13 - Radar Chart:
A radar chart, also known as a spider chart or a star plot, is a graphical representation that displays multivariate data on a two-dimensional polar coordinate system. It is primarily used to compare and visualize the relative values of multiple quantitative variables.

```python
# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [4, 2, 5, 3, 1]

#create a radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(np.linspace(0, 2 * np.pi, len(categories), endpoint=False),
       values, alpha=0.25)

#set category labels
ax.set_xticks(np.linspace(0, 2 * np.pi, len(categories), endpoint=False))
ax.set_xticklabels(categories)

# add title
plt.title('Radar Chart')

#Display the plot
plt.show()
```

# 14 - Tree Map:
A treemap is a visualization technique used to display hierarchical data as a set of nested rectangles. Each rectangle represents a category or node in the hierarchy, and its size is proportional to a specific attribute or value associated with that category. The treemap layout utilizes the area of the rectangles to convey the relative importance or quantity of each category.

```python
import squarify
# Data
sizes = [50, 25, 10, 10]
labels = ['A', 'B', 'C', 'D']

#Create a treemap
squarify.plot(sizes=sizes, label=labels, alpha=0.7)

#ADD title
plt.title('Treemap')

#Display the plot
plt.axis('off')
plt.show()
```

# 16 - Stream Plot:
A stream plot, also known as a streamline plot or vector field plot, is a type of visualization that depicts the flow or movement of particles in a vector field. It is particularly useful for visualizing fluid dynamics, electromagnetic fields, and other continuous vector fields.

```python
#generate random data
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
U = -1 - X**2 + Y
V = 1 + X - Y**2

#Create a Streamplot
plt.streamplot(X, Y, U, V)

#add title
plt.title('Streamplot')

# Display the plot
plt.show()
```

# 17 - Network Graph:
A network graph, also known as a network diagram or graph, is a visual representation of interconnected nodes or entities, often referred to as vertices, and the connections or relationships between them, known as edges. It is used to analyze and visualize complex relationships, networks, and systems.

```python
import networkx as nx
```

```python
#create an empty graph
G = nx.Graph()

#Add nodes
G.add_nodes_from([1, 2, 3, 4, 5])

#add edges
G.add_edges_from([(1,2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 1)])

#Create a network graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)

#add Title
plt.title('Network Graph')

#display the plot
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a network graph
G = nx.Graph()
G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('C', 'D')

# Create a new figure and axes
fig, ax = plt.subplots()

# Set the layout for the network graph
pos = nx.spring_layout(G)

# Draw the network graph on the specified axes
nx.draw(G, pos, with_labels=True, ax=ax)

# Add title
ax.set_title('Network Graph')

# Display the graph
plt.show()
```

# 18 - Sankey plot:
A Sankey plot, also known as a Sankey diagram or flow diagram, is a type of visualization that illustrates the flow of energy, resources, or any other quantity through a system. It depicts the movement or transfer of entities between different stages or categories in a visually intuitive and proportional manner.

```python
from matplotlib.sankey import Sankey
```

```python
# create a sankey diagram
sankey = Sankey()
sankey.add(flows=[0.25, 0.15, 0.4, -0.3, -0.1, -0.1, -0.15],
          labels=['Input 1', 'Input 2', 'Input 3', 'Output 1',
                 'Output 2', 'Output 3', 'Output 4'])
sankey.finish()

#add title
plt.title('Sankey Diagram')

#Dispaly the plot
plt.show()
```

# 20 - Polar Contour Plot:
A polar contour plot is a type of visualization that combines the polar coordinate system with contour lines to represent three-dimensional data. It is used to display data that varies over both the angular and radial dimensions

```python
#Generate random data
theta = np.linspace(0, 2 * np.pi, 100)
r = np.linspace(0, 1, 100)
Theta, R = np.meshgrid(theta, r)
Z = np.sqrt(R)

#Create a polar contour plot
plt.contourf(Theta, R, Z, cmap='viridis')

#add colorbar
plt.colorbar()

#Add title
plt.title('Polar Contour Plot')

#Display the plot
plt.show()
```
